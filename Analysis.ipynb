```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ticker = pd.read_csv('results.csv')
ticker
```

```python
highest= ticker.groupby('Ticker_Name').max()['Max_High'].to_frame()
```

```python
highest
```

Graph which company has the highest price? and what others's highest price in a day.

```python
plt.barh(highest.index,highest.Max_High)
plt.title("11/30/2021 highest price in a day for each company")
plt.xlabel("Price")
plt.ylabel("Comapny")
plt.show()
```

10 companys 11/30/2021 stock price change trend

```python
fig, ax = plt.subplots()
for Ticker_Name, gp in ticker.groupby("Ticker_Name"):
    gp.plot(x="Hour", y="Max_High", ax=ax, label=Ticker_Name,figsize=(10,50),title="trend")
```
